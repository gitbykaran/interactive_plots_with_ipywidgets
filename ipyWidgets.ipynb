# **KDE graph example**

```python
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import clear_output
```

```python
# first we plot a simple kde graph using mpg dataset
cars = sns.load_dataset('mpg')
sns.kdeplot(cars.horsepower,lw=3,fill=True,bw_adjust=1)
plt.xlim(-30,300)
plt.ylim(0,0.03);
```

```python
# to interact with the graph we have to wrap the kde plot in a function
def kde_plot(bw=1,choose='red'):
  sns.kdeplot(cars.horsepower,lw=3,fill=True,bw_adjust=bw,color=choose)
  plt.xlim(-30,300)
  plt.ylim(0,0.03)
  plt.show();

widgets.interact(kde_plot,bw=(.1,3),choose=['red','green','yellow','black']); #describe on which function to operate and what parameter to operate with min and max.
```

## **Animations**

```python
#genrating data
import numpy as np
x = np.linspace(0,2*np.pi,1000)
```

```python
#ploting the sine wave
def sine_wave():
  y=np.sin(x)
  plt.plot(x,y)

sine_wave()
```

```python
# adding a parameter 't' that we will control to the original function
k = np.linspace(0,2*np.pi,1000)
def sine_wave(t=0,freq=1):
  l=np.sin(freq*(k+0.1*t))
  plt.plot(k,l)
  plt.xlim(0,2*np.pi)
  plt.ylim(-1.5,1.5)
  plt.show()

widgets.interact(sine_wave,t=widgets.Play(min=0,max=20),freq=(1,5,0.5)); # widgets.Play() gives us the control of animations.
```

# **Another Example**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.uniform(0,5,size=100)
ep = np.random.normal(size=100)
y=2*x + ep
plt.scatter(x,y)
```

```python
x_values = np.linspace(0,5,1000)
```

```python
#normal slope line
plt.scatter(x,y)
plt.plot(x_values,x_values,lw=3,color='black')
```

# **Creating a Widget**

```python
def slope_line(m=1):
  plt.scatter(x,y)
  plt.plot(x_values,x_values,lw=3,color='black')
  plt.ylim(-1.2,12.0)
  plt.show();

widgets.interact(slope_line,m=(0.2,5,0.2));
```

```python
def slope_line(m=1):
  plt.scatter(x,y)
  plt.plot(x_values,m*x_values,lw=3,color='red')
  plt.ylim(-1.2,12.0);
  plt.show()


widgets.interact(slope_line,m=[0,1,2,3,4]) ;
```

# **Line Visibility and Text**

```python
def slope_line(m=1,line=False,title=''):
  plt.scatter(x,y)
  if line:
    plt.plot(x_values,m*x_values,lw=3,color='red')
  if title:
    plt.title(f"{title}")
  plt.ylim(-1.2,12.0);
  plt.show()


widgets.interact(slope_line,m=[0,1,2,3,4],line=False,title='Give me a title.') ;
```
